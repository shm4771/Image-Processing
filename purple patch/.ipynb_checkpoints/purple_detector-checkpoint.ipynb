```python
#modules
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def read_images_from_folder(folder_name):
    """ returns a list of images in RGB format"""
    """ R: 0-255, G:0-255, B:0-255 """
    images = []
    if os.path.isdir(folder_name):
        for filename in os.listdir(folder_name):
            img_path = os.path.join(folder_name, filename)
            img = cv2.imread(img_path) # default BGR format
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #conversion in RGB
            images.append(img)
        return images
    else:
        raise Exception("folder_name does not exist")
```

```python
def rgb_to_hsv(images=None):
    """ Convertes a list of rgb images to hsv"""
    """ H: 0-179, S:0-255, V:0-255 """
    hsv_images = []
    if images:
        for img in images:
            hsv_image = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
            hsv_images.append(hsv_image)
        return hsv_images
    else:
        raise Exception("empty image list")
```

```python
def apply_mask(images, lower, upper):
    """ apply the purple color mask on a list of images based on upper & lower value for hue and returns a list of masked images"""
    masked_images = []
    if images:
        for img in images:
            mask = cv2.inRange(img, hsv_lower_purple, hsv_upper_purple)
            masked_image = cv2.bitwise_and(img, img, mask=mask)
            masked_images.append(masked_image)
        return masked_images
    else:
        raise Exception("empty image list")
```

```python
##The method used for detecting purple color is : simple rule based thresolding
##We will be using HSV color representation as it seperates the color from saturation & brightness, thus making easier to apply the filter of desired color 
```

```python
# Reading all purple images
purple_images = read_images_from_folder('purple')
plt.imshow(purple_images[0]) #sample
```

```python
# Reading all non purple images
non_purple_images = read_images_from_folder('non_purple')
plt.imshow(non_purple_images[0]) #sample
```

```python
# Converting images from rgb to hsv
purple_images_hsv = rgb_to_hsv(purple_images)
non_purple_images_hsv = rgb_to_hsv(non_purple_images)
```

```python
# In order to find the lower and upper limit for purple color in dataset
#we can plot and see the color distribution in purple and non-purple images
```

```python
#4 first purple images  and their corresponding histograms of hue component
#images
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))
for index, ax in enumerate(axes):
    ax.imshow(purple_images[index])
    
#histograms
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))
for index, ax in enumerate(axes):
    (n, bins, patches) = ax.hist(purple_images_hsv[index][:, :, 0].flatten(), bins=6)
```

```python
#4 first non purple images  and their corresponding histograms of the hue component
#images
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))
for index, ax in enumerate(axes):
    ax.imshow(non_purple_images[index])
    
#histograms
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))
for index, ax in enumerate(axes):
    (n, bins, patches) = ax.hist(non_purple_images_hsv[index][:, :, 0].flatten(), bins=6)
```

```python
## Based on the above images, we can observe that purple images have more pixels in the range 130-170, which is range for purple and pink colors 
#thus thresold for masking
hsv_lower_purple = np.array([130, 0, 0])
hsv_upper_purple = np.array([179, 250, 250])
```

```python
#In order to visualize that the thresold based method will work
# we will create a mask and then apply it on the images 
```

```python
#Apply mask
result_purple = apply_mask(purple_images_hsv, hsv_lower_purple, hsv_upper_purple)
result_non_purple = apply_mask(non_purple_images_hsv, hsv_lower_purple, hsv_upper_purple)
```

```python
#Visualize images
## Press any key to close the window.
## don't close it via mouse else the jupyter kernel will stop working
purple_concat=np.concatenate((result_purple[0],result_purple[1], result_purple[2], result_purple[3]),axis=1)
non_purple_concat=np.concatenate((result_non_purple[0],result_non_purple[1], result_non_purple[2], result_non_purple[3]),axis=1)

cv2.imshow('purple_concat',purple_concat)
cv2.imshow('non_purple_concat',non_purple_concat)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#As we can see that this method will generate false negatives as well as false positives.
#Thus in order to minimize that, an thresold on number of purple pixels can be used to decide image category
```

```python
#calculating upper value for thresold
##Calculating maximum number of pixels falling in range of 130-179 for a non-purple image (false positive)
max_count = float('-inf')
for non_purple_image_hsv in non_purple_images_hsv:
    (n, bins) = np.histogram(non_purple_image_hsv[:, :, 0].flatten(), bins=6, range=[0, 179])
    max_count = max(max_count, n[4]+n[5])
    
print("maximum value of thresold is:", max_count)
purple_thresold_upper = max_count
```

```python
#calculating lower value of thresold
##Finding minimum number of pixels falling in range of 130-179 in a purple image (false negative)
min_count = float('inf')
for purple_image_hsv in purple_images_hsv:
    (n, bins) = np.histogram(purple_image_hsv[:, :, 0].flatten(), bins=6, range=[0, 179])
    min_count = min(min_count, n[4]+n[5])
    
print("minimum vale of thresold is:", min_count)
purple_thresold_lower = min_count
```

```python
#This thresold value can be chosen based on the application
#If higher false postive rate is allowed, we will choose lower value for thresold
#if higher false negative rate is allowed, we will choose higher value for thresold
purple_thresold = (purple_thresold_lower + purple_thresold_upper)//2 # randomly chosen from range for this problem
print("chosen thresold:", purple_thresold)
```

```python
## Final program 
def detect_purple(image_path):
    if os.path.isfile(image_path):
        image = cv2.imread(image_path) # default BGR format
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) #conversion in RGB
        plt.imshow(image)
        image_hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
        (n, bins) = np.histogram(image_hsv[:, :, 0].flatten(), bins=6, range=[0, 179])
        purple_pixels = n[4] + n[5]
        if purple_pixels > purple_thresold:
            print("This image has purple patch")
        else:
            print("This image does not have purple patch")
```

```python
#Test the mode 
#Purple image
detect_purple('purple/14.png')
```
